# Gerando Estatísticas Básicas Sobre os Dados

Quantos municípios possuem cada estado na base de dados

```python
states = df_hosp.Estado.unique()
for state in states:
  print(f'{state} possui informação de {len(df_hosp[df_hosp.Estado == state].Municipio.unique())} municípios')
```

# Estado com Maior Número de Internações por Dengue Anual

```python
years = df_hosp_estados_ano.Ano.unique()
for year in years:
  highest_hosp_admissions = df_hosp_estados.iloc[df_hosp_estados[f'{year}'].argmax()]
  print(f'Estado com maior número de internações em {year}: '
        f'{highest_hosp_admissions.Estado} - {highest_hosp_admissions[f"{year}"]:n}')
```

```python
alt.Chart(df_hosp_estados_ano).mark_line().encode(
    x='Ano',
    y='Internações - Dengue',
    color='Estado',
    strokeDash='Estado',
).interactive()
```

# Município com Maior Número de Internações por Dengue Anual

```python
for year in years:
  highest_hosp_admissions = df_hosp.iloc[df_hosp[f'{year}'].argmax()]
  print(f'Municipio com maior número de internações em {year}: '
        f'{highest_hosp_admissions.Municipio} ({highest_hosp_admissions.Estado}) - '
        f'{highest_hosp_admissions[f"{year}"]:n}')
```

Selecionando os 10 municípios com maior número de internações por dengue em 2019

```python
highest_hosp_admissions = df_hosp.nlargest(10, f'{years[-1]}').Municipio
```

```python
alt.Chart(df_hosp_ano[df_hosp_ano.Municipio.isin(highest_hosp_admissions)]).mark_line().encode(
    x='Ano',
    y='Internações - Dengue',
    color='Municipio',
    strokeDash='Estado',
).interactive()
```

# 10 Municípios com Maior Número de Internações a cada Ano

```python
for year in years:
  highest_hosp_admissions = df_hosp.nlargest(10, f'{year}')
  print(f'\nMunicipios com maior número de internações em {year}: ')
  for i in range(len(highest_hosp_admissions)):
    print(f'\t({highest_hosp_admissions.iloc[i].Estado}) '
          f'{highest_hosp_admissions.iloc[i][f"{year}"]:n}'
          f' \t{highest_hosp_admissions.iloc[i].Municipio}')
```

# Boxplots por Estado

```python
def boxplot(df: pd.DataFrame, x: str, y: str, title: str=''):
  return alt.Chart(df).mark_boxplot(size=50, extent=0.5).encode(
      x=f'{x}:O',
      y=alt.Y(f'{y}:Q', scale=alt.Scale(zero=False)),
      color=alt.Color(x, legend=None)
  ).properties(
      title=title,
      width=800
  ).configure_axis(
      labelFontSize=12,
      titleFontSize=12
  )
```

```python
boxplot(df_hosp_estados_ano, 'Ano', 'Internações - Dengue', 'Internações por Dengue Anual')
```
