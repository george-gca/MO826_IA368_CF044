# Verificando Completude dos Dados

```python
def print_completion_table(df: pd.DataFrame):
  print(f'Total de linhas: {len(df):n}')
  data = []
  for c in df:
    data.append([c, f'{df[c].isna().sum():n}'])
  
  print(tabulate(data, headers=['Coluna', 'Dados Faltantes'], tablefmt='pretty'))
```

```python
print_completion_table(df_hosp)
```

```python
print_completion_table(df_hosp_ano)
```

```python
print_completion_table(df_hosp_estados)
```

```python
print_completion_table(df_hosp_estados_ano)
```

Pretendemos fazer uma análise mais apurada para avaliar individualmente os anos que tiveram aumento/redução e observar padrões, focando em estados como os de maior número de casos (PA e MA).

# Município com Maior Número de Internações por Dengue Anual

```python
for year in years:
  highest_hosp_admissions = df_hosp.iloc[df_hosp[f'{year}'].argmax()]
  print(f'Municipio com maior número de internações em {year}: '
        f'{highest_hosp_admissions.Municipio} ({highest_hosp_admissions.Estado}) - '
        f'{highest_hosp_admissions[f"{year}"]:n}')
```

Como podemos perceber, tirando os anos de 2008 e 2011, em que o município com maior número de casos foi uma capital (Manaus) que possui população elevada, para todos os outros anos os municípios com maior número de internações foram municípios do interior, com provável acesso menor da população a serviços de conscientização e atenção à saúde. Para confirmar tal hipótese será necessário acesso a dados de investimento em conscientização e saneamento básico.

Selecionando os 10 municípios com maior número de internações por dengue em 2019

```python
highest_hosp_admissions = df_hosp.nlargest(10, f'{years[-1]}').Municipio
```

```python
alt.Chart(df_hosp_ano[df_hosp_ano.Municipio.isin(highest_hosp_admissions)]).mark_line().encode(
    x='Ano:O',
    y='Internações - Dengue:Q',
    color='Municipio:N',
    strokeDash='Estado:N',
).interactive()
```

Podemos destacar no gráfico uma variação muito grande dos valores de internações nos municípios de Boa Vista (RR), Cruzeiro do Sul (AC) e Colinas (MA). No caso de Boa Vista, pelo fato de a mudança ter acontecido e a redução ter sido constante, é provável que seja fruto de ações de conscientização e investimento em saúde na região. Nos casos de Cruzeiro do Sul e Colinas, pelo fato de serem valores muito diferentes dos anos anteriores e posteriores, podendo ser erros nos dados ou mesmo um surto de casos.

# 10 Municípios com Maior Número de Internações a cada Ano

```python
for year in years:
  highest_hosp_admissions = df_hosp.nlargest(10, f'{year}')
  print(f'\nMunicipios com maior número de internações em {year}:\n')

  df_highest_per_state = highest_hosp_admissions.groupby('Estado')[f'{year}'].nunique().to_frame()
  df_highest_per_state.index.name = None

  sum_per_state = highest_hosp_admissions.groupby('Estado')[f'{year}'].sum()
  total_per_state = df_hosp[df_hosp.Estado.isin(highest_hosp_admissions.Estado.unique())].groupby('Estado')[f'{year}'].sum()
  percentage = (sum_per_state / total_per_state)
  df_highest_per_state['Porcentagem do Estado'] = percentage

  print(df_highest_per_state, '\n')
  
  for i in range(len(highest_hosp_admissions)):
    print(f'({highest_hosp_admissions.iloc[i].Estado}) '
          f'{highest_hosp_admissions.iloc[i][f"{year}"]:n}'
          f' \t{highest_hosp_admissions.iloc[i].Municipio}')
```

# Boxplots por Estado

```python
def boxplot(df: pd.DataFrame, x: str, y: str, title: str=''):
  return alt.Chart(df).mark_boxplot(size=50, extent=0.5).encode(
      x=x,
      y=alt.Y(y, scale=alt.Scale(zero=False)),
      color=alt.Color(x, legend=None)
  ).properties(
      title=title,
      width=800
  ).configure_axis(
      labelFontSize=12,
      titleFontSize=12
  )
```

```python
boxplot(df_hosp_estados_ano, 'Ano:O', 'Internações - Dengue:Q', 'Internações por Dengue Anual')
```

Nessa imagem está representado o agrupamento de todos os estados por ano, onde podemos observar que nos anos de 2009 a 2013 tivemos uma maior quantidade de internações por dengue, o que está representado tanto pela mediana das caixas como pelos 1o e 3o quartis. Após esse período podemos perceber uma tendência à redução do número de casos, levando a redução tanto da mediana quando dos quartis das caixas, o que não se manteve para o ano de 2019.
