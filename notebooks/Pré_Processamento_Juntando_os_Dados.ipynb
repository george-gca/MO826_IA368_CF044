Municípios que só aparecem em dados de desmatamento

# Unificando os Dados em um Único DataFrame

Removendo municípios que só existem nos dados hospitalares

```python
df_hosp_clean = df_hosp.drop(df_only_in_hosp.index)
print(f'Dados de internação antes: {len(df_hosp):n}')
print(f'Dados de internação depois: {len(df_hosp_clean):n}')
```

Removendo municípios que só existem nos dados de desmatamento

```python
df_desmat_clean = df_desmat.drop(df_only_in_desmat.index)
print(f'Dados de desmatamento antes: {len(df_desmat):n}')
print(f'Dados de desmatamento depois: {len(df_desmat_clean):n}')
```

Ainda há 5 municípios a mais em desmatamento. Provavelmente são os municípios de mesmo nome. Vamos confirmar.

```python
for city in df_desmat_clean[df_desmat_clean.Municipio.duplicated()].Municipio:
  print(f'{df_desmat_clean[df_desmat_clean.Municipio == city][df_desmat_clean.columns[:4]]}\n')
```

De fato, são os municípios de mesmo nome. Removendo os municípios de mesmo nome que não existem nos dados hospitalares e juntando ambos em um único DataFrame.

```python
df_all = df_desmat_clean.merge(df_hosp_clean, on=['Municipio', 'Estado'])
df_all.sort_values(by=['Municipio'], ignore_index=True, inplace=True)
df_all.rename(columns={f'{year}': f'Internacoes{year}' for year in range(2008, 2020)}, inplace=True)
df_all
```

Gerando agora a visualização com anos em uma coluna específica

```python
column_types = ['Desmatado', 'Incremento'] #, 'Floresta', 'Nuvem', 'NaoObservado', 'NaoFloresta', 'Hidrografia', 'PercDesmatado'] #, 'PercIncremento']
```

```python
dfs_alts = []

# doing melt for the first year
column_type = column_types[0]
columns_of_interest = list(df_all.columns[:5]) + [c for c in df_all.columns if c.startswith(column_type)]
print(f'Colunas de interesse: {columns_of_interest}')
df_tmp = df_all[columns_of_interest]
df_tmp = df_tmp.rename(columns={c: c[-4:] for c in columns_of_interest[5:]})
df_tmp = df_tmp.melt(id_vars=columns_of_interest[:5],
        var_name='Ano',
        value_name=column_type)
dfs_alts.append(df_tmp)

# doing melt for the rest of the years
for column_type in column_types[1:]:
  columns_of_interest = ['Municipio'] + [c for c in df_all.columns if c.startswith(column_type)]
  print(f'Colunas de interesse: {columns_of_interest}')
  df_tmp = df_all[columns_of_interest]
  df_tmp = df_tmp.rename(columns={c: c[-4:] for c in columns_of_interest[1:]})
  df_tmp = df_tmp.melt(id_vars=columns_of_interest[:1], 
          var_name='Ano', 
          value_name=column_type)
  
  dfs_alts.append(df_tmp[df_tmp.columns[2:]])

columns_of_interest = ['Municipio'] + [f'Internacoes{year}' for year in range(2008, 2020)]
print(f'Colunas de interesse: {columns_of_interest}')
df_tmp = df_all[columns_of_interest]
df_tmp = df_tmp.melt(id_vars=columns_of_interest[:1], 
        var_name='Ano', 
        value_name='Internações - Dengue')

dfs_alts.append(df_tmp[df_tmp.columns[2:]])

df_all_alt = pd.concat(dfs_alts, axis=1)
```

```python
df_all_alt
```

Aglutinando as informações por estado

```python
df_states = df_all.groupby(['Estado']).sum().reset_index()
columns = list(df_states.columns)
columns = [columns[0]] + columns[3:]
print(f'Deixando somente as colunas:\n{columns}')
df_states = df_states[columns]
```

```python
df_states
```

Gerando agora a visualização com anos em uma coluna específica

```python
dfs_alts = []

column_type = column_types[0]
columns_of_interest = list(df_states.columns[:2]) + [c for c in df_states.columns if c.startswith(column_type)]
print(f'Colunas de interesse: {columns_of_interest}')
df_tmp = df_states[columns_of_interest]
df_tmp = df_tmp.rename(columns={c: c[-4:] for c in columns_of_interest[2:]})
df_tmp = df_tmp.melt(id_vars=columns_of_interest[:2],
        var_name='Ano',
        value_name=column_type)
dfs_alts.append(df_tmp)

for column_type in column_types[1:]:
  columns_of_interest = ['Estado'] + [c for c in df_states.columns if c.startswith(column_type)]
  print(f'Colunas de interesse: {columns_of_interest}')
  df_tmp = df_states[columns_of_interest]
  df_tmp = df_tmp.rename(columns={c: c[-4:] for c in columns_of_interest[1:]})
  df_tmp = df_tmp.melt(id_vars=columns_of_interest[:1], 
          var_name='Ano', 
          value_name=column_type)
  
  dfs_alts.append(df_tmp[df_tmp.columns[2:]])

columns_of_interest = ['Estado'] + [f'Internacoes{year}' for year in range(2008, 2020)]
print(f'Colunas de interesse: {columns_of_interest}')
df_tmp = df_states[columns_of_interest]
df_tmp = df_tmp.melt(id_vars=columns_of_interest[:1], 
        var_name='Ano', 
        value_name='Internações - Dengue')

dfs_alts.append(df_tmp[df_tmp.columns[2:]])

df_states_alt = pd.concat(dfs_alts, axis=1)
```
